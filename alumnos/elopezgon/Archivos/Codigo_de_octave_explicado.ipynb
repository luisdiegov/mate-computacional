```python
for k = 1:6 ##Recorre mi ciclo 6 veces donde k va tomando tomando un valor
  x = linspace(0, 4*pi, 2^(k+1)+1); # Te genera un vector columna de longitud 2^(k+1) +1 en un intervalo de 0 a 4*pi
  subplot(2,3,k), plot(x, cos(3*x)-2*sin(x))#Divide mi impresion en una matriz de dimensiones 2*3, y con k le indico la posicion
    #en ese recuadro se graficara  el vector x en el eje de las abcisas y la funcion cos(3*x)-2*sin(x) en el eje de las ordenadas
  axis([0, 4*pi, -1.5, 1.5])#Establece los limites de los ejes al momento de graficar, es decir, hay un limite inferio de 0 y superior de 
    # 4*pi en el eje de las abcisas y limite inferior y superior de -1.5 y 1.5,respectivamente , en el eje de las ordenadas
    
end#Termina el ciclo;
```
