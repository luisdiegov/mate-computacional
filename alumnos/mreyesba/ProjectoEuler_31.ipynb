```python
from sympy.interactive import printing#importo las funciones necesarias para poder crear polinomios y luego poder expresar la multiplicacion de todos como uno solo
printing.init_printing(use_latex=True)
from sympy import Symbol, expand
```

```python
x = Symbol('x') #tomo el simbolo equis para poder hacer polinomios
mult=1#mult sera mi polinomifinal. La manera en que lo resolvi fue de la forma en que aprendi en algebra superior 2
for i in [1,2,5,10,20,50,100,200]:#con funciones generadoras es facil obtener las combinaciones en la que puedes repartir 200 "dulces" entre 8"niños con siertas restricciones
    pol=0
    j=0
    while (j*i)<201:#ya que pusimos nuestros valores en el rango generamos los polinomios de forma que la separacion entre los terminos es de i exponentes
        pol+=x**(j*i)
        j+=1
    mult*=pol #vamos multiplicando todos los polinomios que vamos generando
pol1=expand(mult)#expresamos todos los polinomios como uno solo
mult
```

```python
pol1.coeff(x**200)#obtenemos el coeficiente de x**200 tal como nos dice la teoria
```
