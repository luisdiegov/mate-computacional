## Pregunta 2

En este ejercicio realizaremos una criba de Eratóstenes

```python
bool_arr = np.ones((100,), dtype=bool)
```

```python
bool_arr[0] = False
bool_arr[1] = False
```

```python
N_max = int(np.sqrt(len(bool_arr))) #Obtiene la raiz cuadrada del número ya que superando este número, los multiplos no primos ya fueron marcados.
for j in range(2, N_max): #Elimina todos los múltiplos de los primeros n números desde 2 hasta N_max
    bool_arr[2*j::j] = False #Marca todos los múltiplos de j como falsos, a partir de 2j dando pasos de j en j
```

```python
bool_arr
```

```python
np.nonzero(bool_arr)
```

```python
def criba_eratostenes(max):
    bool_arr = np.ones((max,), dtype=bool)
    bool_arr[0] = False
    bool_arr[1] = False
    N_max = int(np.sqrt(len(bool_arr)))
    
    for j in range(2, N_max):
        bool_arr[2*j::j] = False 
    
    print np.nonzero(bool_arr)
```

```python
list = [10,100,200]
for i in list: 
    criba_eratostenes(i)
```

## Pregunta 4

(a) Crea una función llamada correcto que reciba un array 1d de tamaño 9 e indique si está completo o no (i.e. incluye los números del 1 al 9.

```python
def correcto(arr):
    
    aux = np.unique(arr)
    if (arr.ndim <> 1 or aux.size <> 9):
        return False  

    
    for i in range(9):
          
        if (aux[i] <> (i+1)):            
            return False
    
    return True
```

(b) Crea una función llamada descomponer3x3 que reciba un array de enteros de 9×9 y lo separe en sus 9 subarreglos 3×3.

```python
def descomponer3x3(mtx):
    if (mtx.ndim <> 2 or np.shape(mtx)[0] <> 9 or np.shape(mtx)[1] <> 9 ):
        return False
    
    ans = []
    
    for i in xrange(0,9,3):
        for j in xrange(0,9,3):
            aux = []
            for x in range(3):
                for y in range(3):
                    aux.append(mtx[i+x][j+y])
            ans.append(np.array(aux))
    
    return ans
    
```

```python
testoksudoku = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2], 
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],

    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],

    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 1]
])
```

```python
descomponer3x3(testoksudoku)
```

(c) Crea la función esSolucion que recibe un array de 9×9 y devuelve si es o no solución del sudoku.

```python
def esSolucion(mtx):
    if (mtx.ndim <> 2 or np.shape(mtx)[0] <> 9 or np.shape(mtx)[1] <> 9 ):
        return False
    
    for i in range (9):
        if not(correcto(mtx[i,:]) and correcto(mtx[:,i])):
            return False
    
    return all(map(correcto, descomponer3x3(mtx)))
```

```python
esSolucion(testoksudoku)
```
