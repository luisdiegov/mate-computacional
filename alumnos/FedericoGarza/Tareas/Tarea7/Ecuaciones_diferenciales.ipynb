## Problema 1

Utilizando el código de runge-kutta de segundo orden como base, implemente el código de runge-kutta de cuarto orden. La función debe de llamarse `rk4`. 

```python
def RK4(y, t, dt, derivadas):
    k0 = dt*derivadas(y, t)
    k1 = dt*derivadas(y + 0.5*k0, t + 0.5*dt)
    k2 = dt*derivadas(y + 0.5*k1, t + 0.5*dt)
    k3 = dt*derivadas(y + k2, t + dt)
    y_next = y + dt*(k0 + 2*k1 + 2*k2 + k3)/6.0
    
    return y_next
```

## Problema 2

La taza a la cual un recurso finito es recogido depende de la dificultad y de la demanda. Supón que la demanda $D$ se incrementa con la cantidad extraída $E$:

$$\dot D = \alpha E$$

Supón que la dificultad de extracción $W$ es inversamente dependiente de la fracción restante del recurso:

$$
W = \frac{1}{1 -E}
$$

Finalmente, supón que la tasa de extracción $R$ es la razón de la demanda a la dificultad.

$$ R = \dot E = \frac{D}{W} $$

Por último, el costo $C$ del recurso depende de la tasa de la extracción y de la demanda:

$$ C = \frac{D}{R}\xi$$,

donde $\xi$ es un factor de escala.


(a) Escribe un programa que dibuje la tasa de la extracción y el costo como una función del tiempo. Usando las condiciones iniciales: $D_0 = 0.1$, $E_0 = 0.1$,  $\alpha = 2.0$ y $\xi = 0.01$. Grafica de 0 a 5 unidades de tiempo. 


```python
def graficar_tasa_y_costo_RK4(alpha, ji, D_0, E_0):
    
    tau = 5
    N = 100
    dt = tau/(float)(N-1) 
    tiempo = np.linspace(0, tau, N)
    r = np.zeros([N,2])
    
    r[0,0] = D_0
    r[0,1] = E_0
    
    def obtencion_recurso(estado, tiempo):
        r0 = alpha*estado[1] # D prima
        r1 = estado[0]*(1-estado[1]) # E prima

        return np.array([r0, r1])
    
    for j in range(N-1):
        r[j+1] = RK4(r[j], tiempo[j], dt, obtencion_recurso)
    
    D = r[:, 0]
    E = r[:, 1]
    
    def tasa_extraccion(demanda, cantidad_extraida):
        return demanda*(1 - cantidad_extraida)

    tasa = np.zeros(N)
    for i in range(N):
        tasa[i] = tasa_extraccion(D[i], E[i])
    
    def costo_extraccion(demanda, tasa):
        return ji*demanda/tasa
    
    costo = np.zeros(N)
    for i in range(N):
        costo[i] = costo_extraccion(D[i], E[i])
    
    fig, ax = plt.subplots(1,2, figsize=(15,8))
    ax[0].plot(tiempo, tasa, color = 'r')
    ax[0].set_title('Tiempo vs tasa',fontsize=10)
    ax[0].set_xlabel('Tiempo (min)',fontsize=10)
    ax[0].set_ylabel(u'Tasa de extracción',fontsize=10)
    ax[1].plot(tiempo, costo, color = 'b')
    ax[1].set_title('Tiempo vs costo',fontsize=10)
    ax[1].set_xlabel('Tiempo (min)',fontsize=10)
    ax[1].set_ylabel(u'Costo de extracción',fontsize=10)
    
```

```python
def graficar_tasa_y_costo_odeint(alpha, ji, D_0, E_0):
    
    tau = 5
    N = 100
    dt = tau/(float)(N-1) 
    tiempo = np.linspace(0, tau, N)
    r = np.zeros([2])
    
    r[0] = D_0
    r[1] = E_0
    
    def obtencion_recurso(estado, tiempo):
        r0 = alpha*estado[1] # D prima
        r1 = estado[0]*(1-estado[1]) # E prima

        return np.array([r0, r1])
    
    solucion = odeint(func = obtencion_recurso, y0 = r, t = tiempo)
    
    D = solucion[:, 0]
    E = solucion[:, 1]
    
    def tasa_extraccion(demanda, cantidad_extraida):
        return demanda*(1 - cantidad_extraida)

    tasa = np.zeros(N)
    for i in range(N):
        tasa[i] = tasa_extraccion(D[i], E[i])
    
    def costo_extraccion(demanda, tasa):
        return ji*demanda/tasa
    
    costo = np.zeros(N)
    for i in range(N):
        costo[i] = costo_extraccion(D[i], E[i])
    
    fig, ax = plt.subplots(1,2, figsize=(15,8))
    ax[0].plot(tiempo, tasa, color = 'r')
    ax[0].set_title('Tiempo vs tasa',fontsize=10)
    ax[0].set_xlabel('Tiempo (min)',fontsize=10)
    ax[0].set_ylabel(u'Tasa de extracción',fontsize=10)
    ax[1].plot(tiempo, costo, color = 'b')
    ax[1].set_title('Tiempo vs costo',fontsize=10)
    ax[1].set_xlabel('Tiempo (min)',fontsize=10)
    ax[1].set_ylabel(u'Costo de extracción',fontsize=10)
```

## Problema 3

(a) Calcular el valor de $\pi$ usando métodos de Monte Carlo.

Este ejemplo lo hicimos en clase; sabiendo que el área de un círculo es $\alpha\times r ²$, determinamos $\alpha$ a través de un círculo de radio $r = 1$.

```python
radio = 1

N = 10000

# Números al azar
X = np.random.uniform(low = -radio, high = radio, size  = N)
Y = np.random.uniform(low = -radio, high = radio, size = N)

# Distancia desde el centro
dist = np.sqrt(X**2+Y**2);

# Aplicamos la fórmula...
caja = (2.0*radio)**2
esta_adentro = dist < radio
N_adentro = np.sum(esta_adentro)
area_circulo = caja*N_adentro/N

# Dibujamos
plt.figure(figsize=(8,8))
plt.scatter(X,Y, c=esta_adentro, s=5.0, edgecolors='none', cmap=plt.cm.Paired)
plt.axis('equal')

print "[Monte Carlo] Área del círculo = ", area_circulo
print "[Exacta] Área del círculo = ", np.pi*radio**2
```

Entre más grande es la $N$, mejor es la aproximación.

(b) El "volumen" de una 2-esfera (un círculo)   $x^2 + y^2 = r^2$ es 

$$\pi r^2$$,

el volumen de la 3-esfera $x^2 + y^2 + z^2 = r^2$ es 


$$\frac{4}{3} \pi r^3$$. 


La 4-esfera $x^2 + y^2 + z^2 + w^2 = r^2$, podemos suponer (usando los ejemplos anteriores) que el hipervolumen de la 4-esfera es 

$$\alpha\pi r^4$$. 


Usando técnicas de Monte Carlo estima $\alpha$.

Para calcular $\alpha$, supongamos $r = 1$; luego el volumen de la 4-esfera es $\alpha\pi$. Denotando por $V_{4}$ el volumen de la 4-esfera, tenemos que $\alpha = \frac{V_{4}}{\pi}$.

Vía montecarlo, calculemos $V_{4}$. El siguiente código se basa en el visto en clase para a 3-esfera de radio $r = 1$:

```python
volumen_hipercubo = 1 * 1 * 1 * 1

N = 1000000
count = 0

puntos = np.array([np.random.uniform(low=0, high=1,size=N), np.random.uniform(low=0, high=1,size=N), 
                   np.random.uniform(low=0, high=1,size=N), np.random.uniform(low=0, high=1,size=N)])  

puntos_adentro = sum(sum(puntos**2) <= 1)

fraccion = 1.0*puntos_adentro/N
volumen = fraccion * volumen_hipercubo
V_4 = volumen*16
```

```python
alpha = V_4 / np.pi
```
