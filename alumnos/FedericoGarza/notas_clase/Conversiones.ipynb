```python
def decimalToBinaryEight(x):
    assert(x>-129 and x<128)
    binario=""
    for i in [7,6,5,4,3,2,1,0]:
        if x/2**i!=0:
            binario=binario+"1"
            x=x%2**i
        else:
            binario=binario+"0" 
    return binario       
```

```python
for i in range(-128,128):
    print str(i)+" "+str(decimalToBinaryEight(i))
```

```python
def binaryEightToDecimal(x):
    decimal=0
    x=str(x)
    if x[0]=='1':
        decimal+=(-2)**7
        for i in range(1, len(x)):
            if x[i]=='1':
                decimal+=2**(7-i) #Necesario el 7-i por como consideramos el rango de i
    else:
        for i in range(0,len(x)):
            if x[i]=='1':
                decimal+=2**(7-i) #Necesario el 7-i por como consideramos el rango de i
    return decimal
```

```python
for i in range(-128,128):
    print str(i)+"="+str(binaryEightToDecimal(str(decimalToBinaryEight(i))))
```
