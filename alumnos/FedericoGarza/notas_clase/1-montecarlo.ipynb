```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from __future__ import division
```

```python
error = np.zeros(1000)
```

```python
radio = 1

for N in range(1, 1001):


# Números al azar
    X = np.random.uniform(low = -radio, high = radio, size  = N)
    Y = np.random.uniform(low = -radio, high = radio, size = N)

# Distancia desde el centro
    dist = np.sqrt(X**2+Y**2);

# Aplicamos la fórmula...
    caja = (2.0*radio)**2
    esta_adentro = dist < radio
    N_adentro = np.sum(esta_adentro)
    area_circulo = caja*N_adentro/N
    
    error[N-1] = np.pi*radio**2 - area_circulo
    
error
# Dibujamos
plt.figure(figsize=(8,8))
plt.plot(np.array([i for i in range(1,1001)]), error)
#plt.axis('equal')
```

```python
# Integrando con Montecarlo
```

```python
def g(x):
    return x**2 + 4*x*np.sin(x)
```

```python
def intg(x):
    return (1/3) * x**3 + 4*(- x*np.cos(x) + np.sin(x))
```

```python
a = 2
b = 5*np.pi

N = 10000

X = np.random.uniform(low=a, high=b, size = N)
Y = g(X)

I_estimado = (b-a)*np.sum(Y) / N;

I_exacto = intg(b) - intg(a)

print "Estimación = ", I_estimado, "Exacto = ", I_exacto
```

```python
(I_exacto - I_estimado)/I_exacto * 100
```

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
plt.figure(figsize = (10,10))

num_particulas = 400

particulas = np.ones([num_particulas,2])*100  # Todas las partículas en el punto (100,100)


# Dibujamos la posición inicial
line, = plt.plot(particulas[:,0], particulas[:,1], 'ro')
plt.xlim(0,200)
plt.ylim(0,200)
```

```python
steps = 5000

for i in range(steps):
    for particula in range(num_particulas):
        particulas[particula, 0] += random.randint(-1, 1)
        particulas[particula, 1] += random.randint(-1, 1)
        
        # Verificar colisión
        x, y = (particulas[particula, 0], particulas[particula, 1])
        if x == 200:
            particulas[particula, 0] = 198
        elif x == 0:
            particulas[particula, 0] = 2
        
        if y == 200:
            particulas[particula, 1] = 198
        elif y == 0:
            particulas[particula, 1] = 2
            
```

```python
# Dibujamos la posición luego de N steps
line, = plt.plot(particulas[:,0], particulas[:,1], 'ro')
plt.xlim(0,200)
plt.ylim(0,200)
```

```python
fig, ax = plt.subplots(1,4, figsize=(17,4))

ax[0]= plt.plot(particulas[0:0,0], particulas[0:0,1], 'ro')

ax[1]= plt.plot(particulas[0:100,0], particulas[0:100,1], 'ro')

ax[2]= plt.plot(particulas[0:200,0], particulas[0:200,1], 'ro')


ax[3]= plt.plot(particulas[0:399,0], particulas[0:399,1], 'ro')
ax[0].xlim(0,200)
ax[0].ylim(0,200)

```
