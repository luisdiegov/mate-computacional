```python
import numpy as np
```

```python
def secante(f, x1, x0, tolerancia, stop):
    
    # Inicializamos el contador
    k = 0
    DF=(f(x0)-f(x1))/(x0-x1)
    # Inicializamos abs(x1-x0) en un número mayor que la tolerancia
    diff = 1
    print "{: >5} {: >10} {: >10} {: >10} {: >11} {: >10}".format("k", "x0", "F(x0)", "DF", "abs(x1-x0)", "x1")
    
    while diff >= tolerancia and k <= stop:
        
        k = k + 1

        x2 = x1 - (float(f(x1)/DF))
                   
        print "{: >5} {: >10.5f} {: >10.5f} {: >10.5f} {: >10.5f} {: >11.5f}".format(k, x0, f(x0), DF, diff, x2)
        
        diff = abs(x2 - x1)
        
        x0 = x1
        x1 = x2
        
    if k > stop:
        print "El método de la secante no convergió: para la tolerancia especificada, se alcanzaron el máximo número de iteraciones"
    
    print "\n\nLa raíz es %2.12f" % x0
```

```python
f = lambda x: (np.sqrt(9.8*x/0.25))*np.tanh(np.sqrt(9.8*0.25/x)*4) - 36
```

```python
secante(f, 70, 69, 10e-8, 500)
```
