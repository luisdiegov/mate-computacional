(e) Encapsula lo que acabamos de hacer en una función llamada criba_eratostenes, que reciba como argumento maximo el cual representa el tamaño del array. Muestra el resultado para maximo = {10, 100, 200}, usando un for.

```python
def criba_eratostenes(maximo):
    array2=np.ones((maximo+1,), dtype=bool)
    array2[0]=False
    array2[1]=False
    stop = int(np.sqrt(len(array2)))
    for j in range (2, stop):
        array2[2*j::j] = False

    return np.nonzero(array2) 
```

```python
criba_eratostenes(10)
```

```python
criba_eratostenes(100)
```

```python
criba_eratostenes(200)
```

## Pregunta 4

El **Sudoku** es un pasatiempo que consiste en un arreglo $9×9$ de celdas, con dígitos del 1 al 9, de tal manera que no se repiten valores en cada columna, cada renglón y en cada subarreglo de $3×3$.

(a) Crea una función llamada correcto que reciba un array $1d$ de tamaño 9 e indique si está completo o no (i.e. incluye los números del 1 al 9.

```python
def correcto(array):
    prueba = np.zeros(9).astype(int)
    if len(array)==9:
        for i in range(9):
            prueba[array[i]-1]=array[i]
    return (prueba == [1,2,3,4,5,6,7,8,9]).all()
            
```

```python
correcto([1,2,4,3,5,8,7,6,9])
```

(b) Crea una función llamada descomponer3x3 que reciba un array de enteros de $9×9$ y lo separe en sus 9 subarreglos $3×3$.

```python
def descomponer3x3(array):
    s1=array[0:3,0:3]
    s2=array[0:3,3:6]
    s3=array[0:3,6:9]
    s4=array[3:6,0:3]
    s5=array[3:6,3:6]
    s6=array[3:6,6:9]
    s7=array[6:9,0:3]
    s8=array[6:9,3:6]
    s9=array[6:9,6:9]
    
    return np.array([s1,s2,s3,s4,s5,s6,s7,s8,s9])
```

```python
arr=np.zeros((9,9)).astype(int)
for i in range (9):
    for k in range (9):
        arr[i,k] = k+1
    arr[i,0]=i+1

        
      
descomponer3x3(arr)
```

(c) Crea la función esSolucion que recibe un array de $9×9$ y devuelve si es o no solución del sudoku.

```python
def esSolucion(array):
    row=0
    col=0
    sub=0
    
    rowA=array
    colA=array.T
    subA=descomponer3x3(array).reshape(9,1,9)
    
    for k in range(9):
        if correcto(rowA[k]):
            row+=1
            
    for j in range(9):
        if correcto(colA[j]):
            col+=1
            
    for i in range(9):
        if correcto((subA[i])[0]):
            sub+=1
    return (row == 9 and col == 9 and sub == 9)
        
            
```

```python
esSolucion(arr)
```

```python
esSolucion(arr2)
```
