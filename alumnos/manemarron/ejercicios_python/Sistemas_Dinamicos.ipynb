```python
%%writefile euler.py
import numpy as np
def euler(y,dy,tau,N):
    dt = tau / float(N-1)
    for t in xrange(N-1):
        y[t+1]=y[t] + dy(y[t])*dt
    return y
```

## Péndulo

```python
%%writefile pendulo.py
import numpy as np
from euler import *

class Pendulo:
    def __init__(self,l,m,g):
        self._m = m
        self._l = l
        self.G = g
        self.frequency = np.sqrt(self.G/self._l)
        self.T = 2*np.pi/self.frequency
        self.dy = lambda y: np.array([y[1],-self.G/self._l*y[0]])
        
    def analytic_position(self,t,theta,omega):
        return (theta*np.cos(t*self.frequency)) + (omega/self.frequency*np.sin(t*self.frequency))
    
    def euler_position(self,y,tau,N):
        return euler(y,self.dy,tau,N)
    
```

```python
from pendulo import Pendulo
```

```python
p = Pendulo(1.0,1.0,9.8)
N=1000
theta = np.zeros(N)
theta[0]=0.2
omega=0.0
time = np.linspace(0,4.0,N)
theta = p.analytic_position(time,theta[0],omega)
```

```python
plt.plot(time,theta)
plt.xlabel("Tiempo")
plt.ylabel("Theta")
```

```python
p = Pendulo(1.0,1.0,9.8)
tau = 5 * p.T 
N = 100
time = np.linspace(0,tau,N)
y = np.zeros([N,2])
y[0]=np.array([0.2,0.0])
theta = np.zeros(N)
theta[0]=0.2
omega=0.0
y = p.euler_position(y,tau,N)
theta = p.analytic_position(time,theta[0],omega)
```

```python
plt.plot(time,y[:,0],label="theta_euler")
plt.plot(time,y[:,1],label="omega")
plt.plot(time,theta,label="theta_analitica")
plt.xlabel("Tiempo")
plt.ylabel("Theta, Omega")
plt.legend(loc="best")
```
