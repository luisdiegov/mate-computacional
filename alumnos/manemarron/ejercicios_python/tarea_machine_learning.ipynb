**(a)** Usando **SimPy** demostrar que 

$$
\frac{\partial}{\partial \beta_j} \textbf{J}(\beta) = \frac{1}{m}\sum_{i=1}^{m}\left(\hat{y}(x^{(i)}) - y(x^{(i)})\right) \cdot x^{(i)}_j
$$

```python
m = Symbol("m")
i = Symbol("i")
x = IndexedBase('x')
y = IndexedBase('y')

J = 1/(2*m)*sym.summation((y[i] - alpha - beta*x[i])**2, (i,1,m))
dJa = sym.simplify(sym.diff(J,alpha))
dJb = sym.simplify(sym.diff(J,beta))
```

## Pregunta 2

Entrenar una regresión lineal. Grafique los datos y el mejor modelo. Explique como llegó a los valores de α. ¿Coinciden con los mostrados en la página web?

```python
chirps_data = read_data("../../../lecture_5/data/chirps.txt")
```

```python
def plotModelo(x,y,rl,alpha=10**-3,):
    modelo = lambda x,b,m: b + m*x # función para graficar el modelo
    
    rl.alpha = alpha
    rl.gradientDescent(chirps_data[:,0],chirps_data[:,1])
    
    _beta = rl.historia['beta'][rl.breaking_iteration-1]

    plt.scatter(x,y, label="datos")
    plt.plot(x, modelo(x, _beta[0], _beta[1]), label="int: %1.2f, pen: %1.2f" % (_beta[0], _beta[1]))
    plt.xlabel('Chirps per second')
    plt.ylabel('Temperature (F)')
    plt.legend(loc="best")
```

```python
rl = RegresionLineal(alpha=10**-3,max_iters=2**100,tols=2**-100)
interact(plotModelo, x=fixed(chirps_data[:,0]), y=fixed(chirps_data[:,1]),rl=fixed(rl),alpha=[10**-3,10**-2,10**-7]);
```

```python
rl = RegresionLineal(alpha=5*10**-3,max_iters=2**100,tols=2**-100)
plotModelo(chirps_data[:,0],chirps_data[:,1],rl,rl.alpha)
```

Utilicé un número de iteraciones muy elevado y una tolerancia bastante pequeña para lograr una buena precisión de los valores de $\beta$. El valor de $\alpha$ en un principio lo había puesto muy grande, y el método daba un overflow. Tras algunas pruebas logré llegar a un conjunto de valores factibles de $\alpha$ entre $10^-3$ y $10^-2$. Luego utilizando el widget fui acercándome más, y finalmente determiné que $5*10^-3$ era el valor más adecuado. La regresión lineal es la misma (redondeada a dos decimales) a la que viene en la página.

## Pregunta 3

**(a)** Usando los datos del [cuarteto de Anscombe](http://en.wikipedia.org/wiki/Anscombe%27s_quartet) Calcule la regresión lineal ¿Qué sucede?

```python
anscombe_data = read_data("data/anscombe.csv", m=8)
```

```python
def plotModeloAnscombe(array,alpha=10**-2,):
    modelo = lambda x,b,m: b + m*x # función para graficar el modelo
    fig, ax = plt.subplots(2,2, figsize=(15,10))
    c = 0
    for i in range(2):
        for j in range(2):
            rl = RegresionLineal(alpha=alpha,max_iters=2**100,tols=2**-100)
            rl.gradientDescent(array[:,2*c],array[:,2*c+1])
            _beta = rl.historia['beta'][rl.breaking_iteration-1]
            ax[i][j].scatter(array[:,2*c],array[:,2*c+1], label="datos")
            ax[i][j].plot(array[:,2*c], modelo(array[:,2*c], _beta[0], _beta[1]), label="int: %1.2f, pen: %1.2f" % (_beta[0], _beta[1]))
            ax[i][j].set_xlabel('X')
            ax[i][j].set_ylabel('Y')
            ax[i][j].legend(loc="best")
            c+=1
```

```python
plotModeloAnscombe(anscombe_data,0.01)
```

Se puede ver en las 4 gráficas que aunque la distribución de los puntos en cada gráfica es muy distinta, la regresión lineal que los aproxima es la misma. Anscombe utilizó este ejemplo para explicar el impacto que tienen puntos distantes al resto de los datos. Estas cuatro distribuciones de datos tienen prácticamente la misma media y varianza muestrales tanto para x como para y. Además, la correlación entre x y y es la misma en las cuatro gráficas.

```python
for i in range(4):
    print "GRAFICA %s:"%(i+1)
    media_x = np.sum(anscombe_data[:,2*i])/len(anscombe_data[:,2*i])
    print "\tMedia muestral de x: %s"%(media_x)
    var_m_x = np.sum((anscombe_data[:,2*i]-media_x)**2)/(len(anscombe_data[:,2*i])-1)
    print "\tVarianza muestral de x: %s"%(var_m_x)
    media_y = np.sum(anscombe_data[:,2*i+1])/len(anscombe_data[:,2*i+1])
    print "\tMedia muestral de y: %s"%(media_y)
    var_m_y = np.sum((anscombe_data[:,2*i+1]-media_y)**2)/(len(anscombe_data[:,2*i+1])-1)
    print "\tVarianza muestral de y: %s"%(var_m_y)
    var_x = np.sum((anscombe_data[:,2*i]-media_x)**2)/(len(anscombe_data[:,2*i]))
    var_y = np.sum((anscombe_data[:,2*i+1]-media_y)**2)/(len(anscombe_data[:,2*i+1]))
    desv_x=np.sqrt(var_x)
    desv_y=np.sqrt(var_y)
    corr = (np.sum((anscombe_data[:,2*i]-media_x)*(anscombe_data[:,2*i+1]-media_y))/(len(anscombe_data[:,2*i])))/(desv_x*desv_y)
    print "\tCorrelación: %s"%(corr)
```
