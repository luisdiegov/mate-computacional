(e) Encapsula lo que acabamos de hacer en una función llamada criba_eratostenes, que reciba como argumento maximo el cual representa el tamaño del array. Muestra el resultado para maximo = {10, 100, 200}, usando un for.

```python
def criba_eratostenes(n):
    bool_arr=np.ones((n+1,), dtype=bool)
    bool_arr[0]=False
    bool_arr[1]=False
    N_max = int(np.sqrt(len(bool_arr)))
    for j in range(2, N_max): 
        bool_arr[2*j::j] = False
    
    arr=np.nonzero(bool_arr)
    print arr
```

```python
criba_eratostenes(10)
```

```python
criba_eratostenes(100)
```

```python
criba_eratostenes(200)
```

## Pregunta 4

El Sudoku es un pasatiempo que consiste en un arreglo 9×9 de celdas, con dígitos del 1 al 9, de tal manera que no se repiten valores en cada columna, cada renglón y en cada subarreglo de 3×3.

(a) Crea una función llamada correcto que reciba un array 1d de tamaño 9 e indique si está completo o no (i.e. incluye los números del 1 al 9.

```python
def correcto(arr):
    arreglo=[1,2,3,4,5,6,7,8,9]
    numeros=0
    j=0
    for i in range(9):
        for j in range(9):
            if arr[i]==arreglo[j]:
                numeros=numeros+1
    if numeros==9:
        return 0
    else:
        return -1
```

```python
correcto([2,1,3,5,4,6,7,8,9])
```

(b) Crea una función llamada descomponer3x3 que reciba un array de enteros de 9×9 y lo separe en sus 9 subarreglos 3×3.

```python
def descomponer3x3(arr):
    arr1=arr.reshape((9,3,3))
    return arr1
```

```python
arr=np.random.rand(9,9)
arr
descomponer3x3(arr)
```

(c) Crea la función esSolucion que recibe un array de 9×9 y devuelve si es o no solución del sudoku.

```python
def esSolucion(arr):
    soluciones=0
    arr1=descomponer3x3(arr)
    for i in range(0,9):
        arr2=arr1[i].reshape(9)
        n=correcto(arr2)
        if n==0:
           soluciones=soluciones+1
    if soluciones==9:
        print "Es solución del sudoku"
    else:
        print "No es solución del sudoku"
        
```

```python
arr=np.array([[[9,2,3,4,6,5,7,8,1],[7,2,6,4,5,3,1,8,9],[1,3,2,5,9,6,7,8,4]],
              [[1,2,3,5,6,4,8,7,9],[2,3,1,4,5,6,7,8,9],[1,2,3,4,5,6,7,8,9]],
              [[3,2,4,1,9,6,7,5,8],[1,4,7,6,5,3,2,9,8],[1,2,4,3,5,9,7,8,6]]])
```

```python
esSolucion(arr)
```
